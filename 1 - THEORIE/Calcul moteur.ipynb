### 1. Calcul du diamètre de la poulie
$$
\text{diamètre} = \frac{\text{module} \times \text{nombre\_dents}}{\pi}
$$

### 2. Calcul du rayon de la poulie
$$
\text{rayon} = \frac{\text{diamètre}}{2}
$$

### 3. Vitesse angulaire en rad/s
$$
\omega = 2\pi f
$$

### 4. Vitesse linéaire
$$
v = \frac{\omega \times r}{\text{ratio}}
$$

### 5. Distance parcourue par step
$$
\text{distance\_par\_step} = \frac{\pi \times \text{diamètre}}{\text{N\_steps}}
$$

### 6. Force maximale disponible
$$
F = \frac{T}{r} \times \text{ratio}
$$

### 7. Masse maximale en axe horizontal
$$
m_{\text{total}} = \frac{F}{a + v \cdot g}
$$


```python
import math
g=9.81

# Exemple d'utilisation
T = 1.0      # Couple du moteur en Nm
a = 30      # Accélération maximale (m/s^2)
v = 2      # Vitesse cible (m/s)
ratio = 1    # Ratio du jeu d'engrenage
N_steps = 200    # Nombre de steps par tour
module = 2    # Module du poulie
nombre_dents = 60    # Nombre de dents du poulie
f_max = 16.666 # Fréquence en Hz
Rpm = f_max * 60

print("Calcul moteur avec comme paramètre :")
print("")
print("Couple moteur : ", T, "Nm")
print("Accélération maximale : ", a, "m/s^2")
print("Vitesse cible : ", v, "m/s")
print("Nombre de steps par tour : ", N_steps)
print("Nombre de dents du poulie : ", nombre_dents)
print("Fréquence maximale : ", f_max, "Hz")
print("Vitesse de rotation: ", Rpm, "Tr/min")
print("-----------------------------")
print("")
print("Pas de coefficient de sécurité, les valeurs sont théoriques")
print("")


# Rayon de la poulie
diametre = round((module * nombre_dents)/math.pi,2)
rayon_poulie = diametre/2
print("Diamètre de la poulie : ", diametre, "mm")
print("Rayon de la poulie : ", rayon_poulie, "mm")


# Vitesse angulaire en rad/s
vitesse_rad_s = (2 * math.pi * f_max)
print(f"Vitesse angulaire : {vitesse_rad_s:.2f} rad/s")

# Vitesse linéaire
vitesse_lin = ( vitesse_rad_s * rayon_poulie ) / ratio
print(f"Vitesse linéaire : {vitesse_lin:.2f} mm/s")


# Distance parcourue par step
distance_par_step = (math.pi * diametre) / N_steps
print(f"Distance parcourue par step : {distance_par_step:.2f}mm")



# Calcul de la force maximale disponible
F = ( T / (rayon_poulie/1000) ) * ratio # Force en newtons (N)
print(f"Force maximale disponible : {F:.2f} N")

# Calcul de la masse maximale
m_total = F / (a + v * g)
m_total = m_total * 1000
print(f"Masse maximale  : {m_total:.2f} g")
```
